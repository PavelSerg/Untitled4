```python
import pandas as pd

df = pd.read_csv (r'company_sales_data.csv')
print (df)
```

```python
total_profit = df['total_profit'].tolist()
total_profit 
```

```python
months = df['month_number'].tolist()
months
```

```python
from matplotlib import pyplot as plt     
plt.plot(months,total_profit,'--',marker='o',markerfacecolor='k',color='green',linewidth=3,label='profit')
plt.title('Year Profit')
plt.xlabel('Months')
plt.ylabel('Months Profit')
plt.legend(loc='lower right')
plt.show()
```

```python
facecream=df['facecream'].tolist()
facewash=df['facewash'].tolist()
```

```python
from matplotlib import pyplot as plt
plt.bar([i-0.2 for i in months],facecream,label='facecream',align='center',width=0.4)
plt.bar([i+0.2 for i in months],facewash,label='facewash',align='center',width=0.4)
plt.xticks(months)
plt.legend(loc='upper right')
plt.title('Face products')
plt.xlabel('Number of months')
plt.ylabel('Number of sold units')
plt.grid(linewidth=1,linestyle="--")
plt.show()
```
